Function signature: write_columns(data,fname)


**Validation Tests** <br>
Check for corner cases and constraints in the inputs enlist all cases used for testing

```
assert isinstance(data, list)
assert isinstance(fname, str)
for num in data:
    assert isinstance(num, int) or isinstance(num, float)
```

**Functional Tests** <br>
Check function output matches expected result enlist all cases used for testing

```
# float case
write_columns([1.1],'result')
f = open("result", "r")
assert f.read() == '1.10,1.21,0.77\n'

# integer case
write_columns([1, 2],'result')
f = open("result", "r")
assert f.read() == '1.00,1.00,0.67\n2.00,4.00,2.00\n'

# empyt case
write_columns([],'result')
f = open("result", "r")
assert f.read() == ''

write_columns([1, -2.1],'result')
f = open("result", "r")
assert f.read() == '1.00,1.00,0.67\n-2.10,4.41,0.77\n'
```
